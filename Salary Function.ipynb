```python
#---------------------------------
# Module 2 Assignment
# Salary Function
#---------------------------------

#Importing necessary libraries
import pandas as pd
import os
import zipfile
```

```python
#importing salary data
#---------

try:
    salary_df = pd.read_csv('C:/Users/amoah/Downloads/archive (2)/Total.csv')
    print("Salary data loaded successfully")
    display(salary_df.head())
except FileNotFoundError:
    print("Error : File not found")


```

```python
#--------------
# Create Employee function
#-----------------------------

def get_worker_details(name):
    try:
        if name in salary_df["EmployeeName"].values:
            worker_data = salary_df[salary_df["EmployeeName"] == name].to_dict(orient="records")[0]
            return worker_data
        else:
            raise ValueError("Worker is not found")
    except Exception as e:
        return str(e)

print(get_worker_details("NATHANIEL FORD"))
```

```python
#-------------------
# Data processing with dictionary
#--------------------------------

Salary_dict = salary_df.set_index("EmployeeName").T.to_dict()

print("\nEmployee details via dictionary")
print(Salary_dict.get("NATHANIEL FORD"))
```

```python
#-----------
# Export Employee details to csv and zip
#-----------------------------

def export_employee_csv(name):
    emp_details = get_worker_details(name)
    # convert to dataframe
    if isinstance(emp_details, dict):
        df = pd.DataFrame([emp_details])
        os.makedirs("Employee Profile", exist_ok=True)
        file_path = f"Employee Profile/{name}.csv"

        # Saving to csv
        df.to_csv(file_path, index=False)

        # Zip file
        with zipfile.ZipFile("Employee_Profile.zip", "w") as zipf:
            zipf.write(file_path)
        print(f"\nEmployee Profile exported and zipped successfully: {file_path}")
    else:
        print(emp_details)
```

```python
export_employee_csv("NATHANIEL FORD")
```
